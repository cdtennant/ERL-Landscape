```python
import matplotlib.pyplot as plt
import math
import numpy as np
import pandas as pd
```

## Data
Use data from Excel spreadsheet compiled after 2017 ICFA ERL Workshop (CERN)

```python
df = pd.read_csv('ERL_Landscape_Data_04102019.csv')
```

```python
df.head(20)
```

```python
#BunchCharge_sqrt = df['BunchCharge'].apply(np.sqrt)
BunchCharge2 = np.power(df['BunchCharge'], 0.75)
```

```python
#plt.scatter(x = gdp_cap, y = life_exp, s = np.array(pop)*2, c = col, alpha = 0.8)
```

```python
df.plot.scatter(x = 'AveCurrent', y = 'Energy', s = BunchCharge2, alpha = 0.75)

plt.xscale('log')
plt.yscale('log')

plt.ylim(1,5000)
plt.xlim(0.0005,150)

xlab = 'Average Current [mA]'
ylab = 'Energy [MeV]'
title = 'ERL Landscape'

plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)

plt.grid(True)
plt.savefig('ERL_landscape.png', dpi=240, bbox_inches='tight')
```
